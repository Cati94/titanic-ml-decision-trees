```python
import numpy as np

array_aleatorio = np.random.randint(0, 101, size=100)
print(array_aleatorio)

countzerofalse=array_aleatorio<=20
countzerotrue=array_aleatorio>20

print("\nNúmeros menores ou iguais a 20 (False):", countzerofalse)
print("\nNúmeros maiores que 20 (True):", countzerotrue)

positions_greater_than_20 = np.where(countzerotrue)[0]
print("\nPosições e Números maiores que 20:")

for pos in positions_greater_than_20:
    print(f"Posição {pos}: Número {array_aleatorio[pos]}")
```
